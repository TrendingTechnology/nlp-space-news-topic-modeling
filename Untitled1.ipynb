````python
# Dataset
publication_name = "guardian"

manual_stop_words = ["nt", "ll", "ve"]

# Data locations
data_dir_path = str(
    Path().cwd() / "data" / "processed" / f"{publication_name}_processed.csv"
)

minibatch = False
````

````python
# Get stop words from all packages
# NLTK
if not ((Path.cwd().parents[1]) / "nltk_data").exists():
    nltk.download("punkt")
    nltk.download("wordnet")
    nltk.download("stopwords")
    nltk.download("averaged_perceptron_tagger")
nltk_stop_words = set(stopwords.words("english"))
# Spacy and sklearn
spacy_stop_words = STOP_WORDS
sklearn_stop_words = stop_words.ENGLISH_STOP_WORDS

# Assemble manual list of stop words
spacy_not_in_sklearn = set(spacy_stop_words) - set(sklearn_stop_words)
nltk_not_in_sklearn = set(nltk_stop_words) - set(sklearn_stop_words)
all_stop_words = set(
    list(set(sklearn_stop_words))
    + list(spacy_not_in_sklearn)
    + list(nltk_not_in_sklearn)
)

# Manually add to stop words
for manual_stop_word in manual_stop_words:
    all_stop_words.add(manual_stop_word)
````

````python
# Preprocessing and tokenizing
def preprocessing(line):
    line = line.lower()
    line = re.sub(r"[{}]".format(string.punctuation), " ", line)
    return line
````

````python
df = pd.read_csv(Path(data_dir_path))
df = df[["text", "year"]]
display(df.head())
````

````python
corpus_raw = df.loc[:, "text"].values.tolist()
````

````python
cell_st = time()

vectorizer = TfidfVectorizer(
    tokenizer=None,
    preprocessor=preprocessing,
    stop_words=all_stop_words,  # "all_stop_words" or "english"
    lowercase=True,
    ngram_range=(1, 1),
    max_df=1.0,
    min_df=1,
    max_features=None,
    binary=False,
    strip_accents="ascii",
    token_pattern="[a-z][a-z]+",
)
# X = vectorizer.fit_transform(corpus_raw)

total_minutes, total_seconds = divmod(time() - cell_st, 60)
print(
    f"Cell exection time: {int(total_minutes):d} minutes, {total_seconds:.2f} seconds"
)
````

````python
cell_st = time()

if minibatch:
    km = MiniBatchKMeans(
        n_clusters=15,
        init="k-means++",
        n_init=1,
        init_size=1000,
        batch_size=1000,
        verbose=opts.verbose,
    )
else:
    km = KMeans(
        n_clusters=15,
        init="k-means++",
        max_iter=100,
        n_init=1,
        verbose=False,
        random_state=42,
    )
pipe = Pipeline(
    steps=[
        ("cleaner", TextCleaner(split=False)),
        ("vectorizer", vectorizer),
        ("kmeans", km),
    ]
)
doc_topic = pipe.fit_transform(corpus_raw)

total_minutes, total_seconds = divmod(time() - cell_st, 60)
print(
    f"Cell exection time: {int(total_minutes):d} minutes, {total_seconds:.2f} seconds"
)
````

````python
Xtr = pipe.named_steps["vectorizer"].transform(corpus_raw)
print(
    "Silhouette Coefficient: %0.3f"
    % silhouette_score(Xtr, pipe.named_steps["kmeans"].labels_, sample_size=df.shape[0])
)

print("Top terms per cluster:")
order_centroids = pipe.named_steps["kmeans"].cluster_centers_.argsort()[:, ::-1]
terms = pipe.named_steps["vectorizer"].get_feature_names()
for i in range(15):
    print("Cluster %d:" % i, end="")
    for ind in order_centroids[i, :10]:
        print(" %s" % terms[ind], end="")
    print()
````

````python
def find_optimal_clusters(pipe, data, k_min, k_max, minibatch=False):
    iters = range(k_min, k_max + 1)

    sse = []
    for k in iters:
        if minibatch:
            km = MiniBatchKMeans(
                n_clusters=k,
                init="k-means++",
                n_init=1,
                init_size=1000,
                batch_size=1000,
                verbose=opts.verbose,
            )
        else:
            km = KMeans(
                n_clusters=k,
                init="k-means++",
                max_iter=100,
                n_init=1,
                verbose=False,
                random_state=42,
            )
        pipe = Pipeline(
            steps=[
                ("cleaner", TextCleaner(split=False)),
                ("vectorizer", vectorizer),
                ("kmeans", km),
            ]
        )
        start_time = time()
        pipe.fit(data)
        cluster_inertia = pipe.named_steps["kmeans"].inertia_
        time_spent = time() - start_time
        sse.append(cluster_inertia)
        print(
            f" > Fit {k} clusters and found inertia = {cluster_inertia:.2f}, "
            f"in {time_spent:.2f} seconds"
        )

    f, ax = plt.subplots(1, 1)
    ax.plot(iters, sse, marker="o")
    ax.set_xticks(iters)
    ax.set_xticklabels(iters)
    ax.set_title("SSE by Cluster Center Plot", fontweight="bold", loc="left")


find_optimal_clusters(pipe, corpus_raw, 10, 20)
````

````python
X_dist = pipe.named_steps["kmeans"].fit_transform(Xtr) ** 2
df["sqdist"] = pd.Series(X_dist.sum(axis=1).round(2))
display(df.head(30))
````

````python
# Get indices of points for each cluster
mydict = {i: np.where(pipe.named_steps["kmeans"].labels_ == i)[0].tolist() for i in range(pipe.named_steps["kmeans"].n_clusters)}
df["most_popular_topic"] = np.nan
for k, v in mydict.items():
    df.loc[v, "most_popular_topic"] = mapping_dict[publication_name][k]
````

````python
topics_by_timeframe = (
    df.groupby(["most_popular_topic", "year"])
    .size()
    .reset_index()
    .sort_values(by=["most_popular_topic", 0, "year"], ascending=False)
    .rename(columns={0: "count"})
)
display(topics_by_timeframe.head())
````
